### Problem Sentence
The Brown University Standards Corpus of Present-Day American English also known popularly as Brown Corpus was compiled in the 1960s as a general corpus in the field of corpus linguistics. It contains 500 samples of English- language text, totaling roughly one million words, compiled from work published in the United States in 1961.

```python
import nltk
```

```python
with open('brown_corpus_ca10.txt','r') as myfile:
    data = myfile.read().replace('\n','')
```

```python
data2 = data.replace('/','')
```

```python
from nltk import sent_tokenize,word_tokenize
```

```python
sent_tokenize(data2)
```

```python
for sent in sent_tokenize(data2):
    print(word_tokenize(sent))
```

```python
from nltk.corpus import stopwords
stopwords_eng = set(stopwords.words('english'))
```

```python
single_tokenized_lower = list(map(str.lower, word_tokenize(data2)))
print(single_tokenized_lower)
```

```python
print([word for word in single_tokenized_lower if word not in stopwords_eng])
```

```python
from string import punctuation
print(type(punctuation), punctuation)
```

```python
stopwords_eng_withpunct = stopwords_eng.union(set(punctuation))
print(stopwords_eng_withpunct)
```

```python
filtered_data2 = [word for word in single_tokenized_lower if word not in stopwords_eng_withpunct]
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
stemmed_data2 = [porter.stem(word) for word in filtered_data2]
print(stemmed_data2)
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
lemmatized_data2 = [wnl.lemmatize(word) for word in stemmed_data2]
print(lemmatized_data2)
```

```python
for i in sent_tokenize(data2):
    wordlist = nltk.word_tokenize(i)
    wordlist = [w for w in wordlist if not w in stopwords_eng_withpunct]
    tagged = nltk.pos_tag(wordlist)
    print(tagged)
```

```python
sentences = sent_tokenize(data2)
tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
tagged_sentences = [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
chunked_sentences = nltk.ne_chunk_sents(tagged_sentences, binary=True)

def extract_entity_names(t):
    entity_names = []

    if hasattr(t, 'label') and t.label:
        if t.label() =='NE':
            entity_names.append(' '.join(child[0] for child in t))
        else:
            for child in t:
                entity_names.extend(extract_entity_names(child))
    return entity_names

entity_names = []
for tree in chunked_sentences:
    entity_names.extend(extract_entity_names(tree))
print(set(entity_names))
```
