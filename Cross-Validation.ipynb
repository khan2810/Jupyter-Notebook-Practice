### Problem Statement
In this analysis, we will generate the Random Forest Using the Cros Validation spiltting technique instead of train_test_split using Iris dataset.

```python
from sklearn.datasets import load_iris
iris_data = load_iris()
print(iris_data)
```

```python
inp = iris_data.data
out = iris_data.target
print(inp)
```

```python
print(out)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True)
```

```python
print("Train Set             Test Set")
for trainset,testset in kf.split(inp):  # Returning the index
    print(trainset,testset)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(rf,inp,out,scoring='accuracy',cv=10))
```

```python
acc = cross_val_score(rf,inp,out,scoring='accuracy',cv=10).mean()*100
print("Accuracy of the model is ", acc)
```
